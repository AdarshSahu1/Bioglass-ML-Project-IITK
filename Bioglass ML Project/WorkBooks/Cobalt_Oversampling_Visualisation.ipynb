```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.multioutput import MultiOutputRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load dataset
file_path = r"C:\Users\29200\Downloads\Dataset\Co_oversampled_resampled.xlsx"
df = pd.read_excel(file_path)

# Prepare X and y
X = df[['Conc.', 'Cell Viability 24', 'Cell Viability 48', 'Cell Viability 72', 
         'ALP 7', 'ALP 14','Surface area m2/g', 'Pore volume cm3/g', 'Pore size nm']]
y = df[['SiO2', 'B2O3', 'CaO', 'Na2O', 'P2O5', 'Co','CoO']]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.28, random_state=42)

# Choose a multioutput regression model (Example: MultiOutputRegressor with KNeighborsRegressor)
model = MultiOutputRegressor(KNeighborsRegressor(algorithm='ball_tree', n_neighbors=4, weights='distance', n_jobs=-1))

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate RMSE and R² score for each output variable
overall_rmse = 0
fig, axs = plt.subplots(2, 4, figsize=(16, 8))  # 2 rows, 4 columns for 8 variables

for i, column in enumerate(y.columns):
    rmse = np.sqrt(mean_squared_error(y_test.iloc[:, i], y_pred[:, i]))
    r2 = r2_score(y_test.iloc[:, i], y_pred[:, i])
    overall_rmse += rmse
    print(f"{column}: RMSE = {rmse:.4f}, R² = {r2:.4f}")
    
    # Plotting actual vs predicted
    row = i // 4
    col = i % 4
    axs[row, col].scatter(y_test.iloc[:, i], y_pred[:, i], color='blue', alpha=0.7)
    axs[row, col].plot([y_test.iloc[:, i].min(), y_test.iloc[:, i].max()], 
                       [y_test.iloc[:, i].min(), y_test.iloc[:, i].max()], 
                       color='red', linestyle='--')  # y = x line
    axs[row, col].set_title(f"Actual vs Predicted for {column}")
    axs[row, col].set_xlabel("Actual Values")
    axs[row, col].set_ylabel("Predicted Values")
    axs[row, col].grid(True)

# Adjust layout
plt.tight_layout()

# Save the plot as .png file
plt.savefig(r"C:\Users\29200\Downloads\Dataset\Co_Results\actual_vs_predicted_matrix.png")

# Show the plot (optional)
plt.show()

# Calculate overall RMSE
```
